```python
import pandas as pd
import numpy as np
from reliability.Distributions import Weibull_Distribution
import matplotlib.pyplot as plt
```

```python
a = [88.25, 90.833, 100, 105.833, 121.666, 53.25, 72.5, 89.25, 101.75, 118.25, 121.666, 93.25]
```

```python
l = 0.008
t = 0.0016
w = 0.003

b = []

for i in a:
    b.append((i*l)/(w*t*t))
```

```python
df = pd.DataFrame()
df['a'] = a
df['b'] = b
```

```python
df = df.sort_values('a')
```

```python
p = []

for i in range(len(df['a'])):
    p.append((i+1)/(13))
```

```python
y = []

for i in p:
    y.append(np.log(np.log(1/(1-i))))
```

```python
x = []

for i in df['b']:
    x.append(np.log(i))
```

```python
res = np.polyfit(x,y,1)
res
```

```python
sig = np.exp(-res[1]/res[0])
sig
```
